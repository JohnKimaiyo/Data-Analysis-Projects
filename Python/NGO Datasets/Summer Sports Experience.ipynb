# 3 What was the highest month for Attendance? How much was number of Attendance  that month?

```python
# Load in some packages
import calendar
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter

warnings.filterwarnings("ignore")

# 1. What was the highest month for spillage? How much was the total for  that month?

# sport_dfReplace NaN or inf values in the 'Month' column with a default value (e.g., 0)
sport_df['Month'] = sport_df['Month'].fillna(0).astype(int)

# Convert month numbers to abbreviated month names
sport_df['Month Name'] = sport_df['Month'].apply(lambda x: calendar.month_abbr[x])

# Group by month and calculate total attendance  for each month
attendamce_by_month = sport_df.groupby('Month Name').sum()['Attendance Sum']

# Find the best month for spilage  and the corresponding earnings
highest_month = attendamce_by_month.idxmax()
attendance_for_highest_month = attendamce_by_month.max()

print(f"The highest month for attendance was {highest_month} with attendance number of {attendance_for_highest_month :,.2f}")
```

# 3 Which Borough Locations had the highest number of Attendance ?


```python
# Plot the highest number of spilage for each locality
# Load in some packages
import calendar
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter


attendance_by_location = sport_df.groupby('Borough Location').sum()['Attendance Sum'] # Take the number of Attendance for each Locality

# Sort the values in descending order and select the top five
top_five_locations = attendance_by_location.sort_values(ascending=False).head(5)

# Display the result
print(top_five_locations)

# Plot the top Five Localities
top_five_locations.plot(kind='bar', color='purple', figsize=(10, 6))
plt.title('Top Five Borough Location with Most Attendance')
plt.xlabel('Borough Location')
plt.ylabel('Attendance Count')
plt.show()

```

# 4 Which Park Location had the highest number of Attendance?

```python
attendance_by_location = sport_df.groupby('Park Location').sum()['Attendance Sum'] # Take the number of Attendance for each Locality

# Sort the values in descending order and select the top five
top_five_Park_Location = attendance_by_location.sort_values(ascending=False).head(5)

# Display the result
print(top_five_Park_Location)

# Plot the top Five Localities
top_five_Park_Location .plot(kind='bar', color='green', figsize=(10, 6))
plt.title('Top Five Park Location with Most Attendance')
plt.xlabel('Park Location')
plt.ylabel('Attendance Count')
plt.show()

```
